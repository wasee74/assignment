```python
import numpy as np
import pandas as pd
```

```python
#Reading the file
dataframe1=pd.read_csv(r"C:\\Waseem\\Cambridge NIAB.csv", low_memory=False)
```

```python
from plotnine import ggplot, aes, geom_line
```

```python
dataframe2=dataframe1[dataframe1['Rain'].notna()]
ggplot(dataframe2) + aes(x="Mon", y="Rain") + geom_line()
```

```python
from plotnine import geom_point
```

```python
dataframe3=dataframe2[dataframe2['Sun_Duration'].notna()]
ggplot(dataframe3) + aes(x="Mon", y="Sun_Duration") + geom_point()
```

```python
from plotnine import labs
```

```python
dataframe4=dataframe1[dataframe1['Tmax'].notna()]
(
    ggplot(dataframe4)
    + aes(x="Mon", y="Tmax", color="Mon")
    + labs(
        x="Month",
        y="Maximum Temperature",
        color="Month",
        title="Maximum Temperature in different months",
    )
    + geom_point()
)
```
